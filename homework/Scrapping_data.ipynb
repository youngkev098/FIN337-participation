```python
import requests 

base_url   = 'http://chroniclingamerica.loc.gov/search/pages/results/'
parameters = '?andtext=lynching&format=json'

url = base_url + parameters
print(url)

r = requests.get(url) # runs search 

```

```python
r.status_code # 200 hundred means good. 404 error code the internet is bad 
```

```python
r.json() # json is basically like a dictionary 
```

```python
search_json = r.json()
search_json.keys() # get a list of elements 
search_json['totalItems']
```

```python
type(search_json['items'])
# len(search_json['items'])
type(search_json)
```

```python
search_json['items'][0]['county']
search_json['items'][0].keys()
search_json['items'][0]['title']
```

```python
# pandas can import json and a url into a DF 
pd.DataFrame(search_json['items'])
```
