```python
import pandas as pd
# this "firm" variable is really "last firm" but the database manager called it what they called it 
# and now IT is scared that fixing it will break all their systems, so it's "stuck as-is"
some_employees = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue','Austin'],
                    'firm': ['Citi','GS','Accenture','Wells Fargo','Lehigh'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR','Finance'],
                    'hire_date': [2004, 2008, 2012, 2014,2020]})

departments = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR', 'Janitors'],
                    'employees': [100,200,1,2] })

current_employer = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'firm': ['Yahoo','Yahoo','Yahoo','Yahoo']})

some_employees_alt = pd.DataFrame({'employee': ['Jonathan', 'Mary', 'Lisa', 'Sue'],
                    'firm': ['?','Lehigh','Accenture','Wells Fargo'],
                    'group': ['Accounting', 'CEO', 'Engineering', 'HR'],
                    'hire_date': [2010, 2019, 2012, 2014]})
```

```python
#always specifc how and siffix when merging 
# some_employees.merge(departments,on='group',how='inner')
pd.merge(left=some_employees, right=departments, how='inner', on='group')
# pd.merge(left=some_employees, right=departments, how='left', on='group')
```

```python
pd.merge(left=some_employees, right=departments, how='right', on='group')
```

```python
pd.merge(left=some_employees, right=departments, how='outer', on='group')
```

```python
# pd.merge(left=some_employees, right=current_employer, how='inner', on='employee')
#renaming colums before 
pd.merge(left=some_employees, right=current_employer, how='inner', on='employee', 
         suffixes=('_prior', '_current'))

#renaming colums after the fact 
a =pd.merge(left=some_employees, right=current_employer, how='inner', on='employee')
a=a.rename(columns = {'firm_x': 'firm_prior',
                    'firm_y': 'firm_current'   })

```

# Dealing With Missing Values 

```python
# can interporlate missing vlaues --> smooth the gap
import pandas as pd
import numpy as np

df = pd.DataFrame({"A":[12, 4, 5, None, 1], 
                   "B":[None, 2, 54, 3, None], 
                   "C":[20, 16, None, 3, 8], 
                   "D":[14, 3, None, None, 6]}) 
```

```python
df
```

```python
df.fillna(0) # filled all values 
df2 = df.copy()
df2['A'] = df2['A'].fillna(0) #fill only one column
df2
```

```python
df.fillna(df.mean()) # can fill all missing values with mean, median etc
```

```python
#fillna method parameter 
df.fillna(method='ffill')
```

```python
df4  = df.copy()
df4['firm'] = 1 

df5 = df.copy()
df5['firm'] = 2 

df6 = df4.append(df5).reset_index()
df6
```

# Fill forward separetly to each firm 

- Important thing to note here is that .apply (does something. if the 
    thing you want to do is not a function like sum, use a lambda)


```python
# df6.groupby('firm').fillna(method='ffill')
df6.groupby('firm').apply(lambda x: x.fillna(method='ffill')) 
```
