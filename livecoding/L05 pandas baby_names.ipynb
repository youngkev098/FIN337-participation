```python
somenames = baby_names.loc[baby_names['name'] == 'John']
# print(somenames)
combined = somenames.groupby(['name', 'year']) ['count'].sum().reset_index() 
# reset index turns year from index into a variable
sns.lineplot(data = combined, x = 'year', y = 'count')
```

```python
# look at other names
# hue = name does plotting thing seperately for each name'

names_i_want = (baby_names['name'] == 'Sansa') | (baby_names['name'] == 'Arya')
somenames = baby_names.loc[(baby_names[names_i_want])]

combined = somenames.groupby(['name', 'year']) ['count'].sum().reset_index() # reset index turns year into a variable
sns.lineplot(data = combined, hue= 'name', x = 'year', y = 'count')
```

```python
#Query !!!
somenames = baby_names.query('name in ["Araya", "Sansa"]')
combined = somenames.groupby(['name', 'year']) ['count'].sum().reset_index()
sns.lineplot(data = combined, x = 'year', y = 'count', hue = 'name')
```
